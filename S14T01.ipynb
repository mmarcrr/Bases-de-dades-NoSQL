```python
df_avg = pd.DataFrame(list(db.llum.aggregate(
    [
        {
            "$group": { "_id":"$hour", "priceAverage":{"$avg":"$price"}}
            
        },
        {
            "$sort": { "_id": 1 }
        }
    ]
)))
df_avg
```

```python
plt.figure(figsize=(24,10))
df_avg.plot(x='_id', y='priceAverage', kind='line', ylabel='price', xlabel='hour', title='mitja cost hora')
```

```python
df = pd.DataFrame(db.llum.find())
df.drop(["_id"],axis=1,inplace=True) 
```

```python
df.info()
```

```python
def cat_std_summary(df, column_name,):

  column = df[column_name]
  num_values = column.value_counts().shape[0]
  count_value = column.count()
  freq = column.value_counts(normalize=True)*100
  return [column_name, count_value, num_values, freq]



categorical_columns=list(df.select_dtypes(include=['object']).columns)
table = []
for column in categorical_columns:
  table.append(cat_std_summary(df,column))
print(tabulate(table, headers=["columna","num_Valors","valors","freq %"]))
```
